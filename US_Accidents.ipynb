```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
file_name = './us-accidents/US_Accidents_March23.csv'
df = pd.read_csv(file_name)
```

```python
cities_by_accidents = df['City'].value_counts()
cities_by_accidents[:20]
```

Top 20 Cities with most Accidents

```python
cities_by_accidents[:20].sort_values(ascending = True).plot(kind = 'barh')
plt.xlabel('Accidents')
plt.title('Cities By Accidents')
```

```python
states_by_accidents = df['State'].value_counts()
states_by_accidents.head()
```

Top 20 States with most accidents

```python
states_by_accidents[:20].sort_values(ascending = True).plot(kind = 'barh', xlabel = 'Accidents', title = 'states_by_accidents')
```

```python
df.columns
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='ISO8601')
```

```python
df['Start_Time'].dt.hour.value_counts().head()
```

Hourly Accident Distribution

```python
sns.histplot(df['Start_Time'].dt.hour, bins = 24)
plt.xlabel('Hours')
plt.ylabel('No. of Accidents')
plt.title('Hourly Accident Distribution')
```

--------------------------------------------------------------------------

Weekly Distribution of Accidents

```python
df['Start_Time'].dt.dayofweek.value_counts()
```

```python
sns.histplot(df['Start_Time'].dt.dayofweek, bins = 7)
plt.xlabel('Days of The Week')
plt.ylabel('No. of Accidents')
plt.title('Weekly Accident Distribution')
```

---------------------------------------------------------------------

Hourly Accident Distribution During the Weekends

Saturday

```python
saturday_start_time = df['Start_Time'][df['Start_Time'].dt.dayofweek == 5]
sns.histplot(saturday_start_time.dt.hour, bins = 24)
plt.xlabel('Hours')
plt.ylabel('No. of Accidents')
plt.title('Saturday Hourly Accident Distribution')
plt.style.use('tableau-colorblind10')
```

Sunday

```python
sunday_start_time = df['Start_Time'][df['Start_Time'].dt.dayofweek == 6]
sns.histplot(sunday_start_time.dt.hour, bins = 24)
plt.xlabel('Hours')
plt.ylabel('No. of Accidents')
plt.title('Sunday Hourly Accident Distribution')
plt.style.use('tableau-colorblind10')
```

---------------------------------------------------------------

Monthly Accident Distribution

```python
sns.histplot(df['Start_Time'].dt.month, bins = 12)
plt.xlabel('Months')
plt.ylabel('No. of Accidents')
plt.title('Montly Accident Distribution')
plt.style.use('tableau-colorblind10')
```

----------------------------------

Yearly Accident Distribution

```python
df['Start_Time'].dt.year.value_counts().sort_index(ascending = True)
```

```python
sns.histplot(df['Start_Time'].dt.year, bins = 8)
plt.xlabel('Year')
plt.ylabel('No. of Accidents')
plt.title('Yearly Accident Distribution')
plt.style.use('tableau-colorblind10')
```
